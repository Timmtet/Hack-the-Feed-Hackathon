### a. Update the Post Type column

We'll enhance the 'Post Type' column to accurately represent the type of post made in each network. If the 'Post Type' is not 'Tweet', we'll prefix it with the respective network identifier for clarity

```python
# Define a function to modify 'Post type' based on conditions
def modify_post_type(network, post_type):
    if post_type != 'Tweet':
        return f'{network} {post_type}'
    else:
        return post_type

# Apply the function to create a new modified 'Post type' column
merged_df['Post Type'] = merged_df.apply(lambda x: modify_post_type(x['Network'], x['Post Type']), axis=1)
```

**Even after removing columns entirely filled with nulls, a notable subset of columns still contains a significant proportion of null values in comparison to the actual data. These columns, with a high percentage of nulls, might not provide substantial value for the analysis**

Next, let's develop a function to gain a comprehensive understanding of the updated dataset, presenting the percentage of missing values for each column

```python
def missing_zero_values_table(df):
        zero_val = (df == 0.00).astype(int).sum(axis=0)
        mis_val = df.isnull().sum()
        mis_val_percent = 100 * df.isnull().sum() / len(df)
        mz_table = pd.concat([zero_val, mis_val, mis_val_percent], axis=1)
        mz_table = mz_table.rename(
        columns = {0 : 'Zero Values', 1 : 'Missing_Values', 2 : '% of Total Values'})
        mz_table['Total Zero Missing Values'] = mz_table['Zero Values'] + mz_table['Missing_Values']
        mz_table['% Total Zero Missing Values'] = 100 * mz_table['Total Zero Missing Values'] / len(df)
        mz_table['Data Type'] = df.dtypes
        return mz_table

df_missing_vals = missing_zero_values_table(df_new)
```

### g. Engineering the 'Time of day' column

Now, from the time column, we will engineer another column called 'Time of day'. this will represent the time of the day a post was made

```python
# Define a function to determine the time of day
def get_time_of_day(time_str):
    time = pd.to_datetime(time_str).time()
    if time >= pd.to_datetime('12:00:00').time() and time <= pd.to_datetime('18:00:00').time():
        return 'Afternoon'
    elif time >= pd.to_datetime('18:00:01').time() and time <= pd.to_datetime('23:59:59').time():
        return 'Evening'
    else:
        return 'Morning'

# Apply the function to create the 'Time of day' column
df_new['Time of day'] = df_new['Time'].apply(get_time_of_day)
```

### h. Engineering the engagements score column

To comprehensively measure the engagement of a post, we'll introduce a new metric named 'Engagement score'. This metric represents the weighted average of the engagement components such as likes, comments, shares, and more, that a post accrues. It serves as an independent metric to thoroughly analyze the overall engagement of a post.

Note: We will assign weights to the features, emphasizing human interactions such as Reactions, Likes, Comments, and Shares

```python
# List of numeric features to consider
numeric_features = ['Reactions', 'Likes', 'Comments', 'Shares', 'Impressions', 
                     'Organic Impressions', 'Engagement Rate (per Impression)', 
                     'Engagements', 'Click-Through Rate', 'Post Clicks (All)']

# Assign the specified weights to each feature
weights = {
    'Reactions': 2,
    'Likes': 2,
    'Comments': 2,
    'Shares': 2,
    'Impressions': 0.3,
    'Organic Impressions': 0.3,
    'Engagement Rate (per Impression)': 0.3,
    'Engagements': 0.3,
    'Click-Through Rate': 0.4,
    'Post Clicks (All)': 0.4
}

# Calculate the weighted average for each row
df_new['Engagement score'] = df_new[numeric_features].mul(weights).sum(axis=1) / sum(weights.values())
```

742 posts had a score above 1000

Next, we'll aggregate the text from the 'text' column to form a cohesive body of text. This consolidation allows us to analyze the content of all posts collectively.

```python
# Combine all content in the 'Post' column into a single text
combined_text_high = ' '.join(df_high['Post'])
```

```python
combined_text_high
```

Great, we have the texts together now

Now, we will import the required libraries for the text processing. After which we will normalize, tokenize and lemmatize the text

```python
# import the required dependencies
import re
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
def tokenize(text):
    # normalize case and remove punctuation
    text = re.sub(r"[^a-zA-Z0-9]", " ", text.lower())
    
    # tokenize text
    tokens = word_tokenize(text)
    
    # define lemmatizer
    lemmatizer= WordNetLemmatizer()
    
    # lemmatize and remove stop words
    tokens = [lemmatizer.lemmatize(word) for word in tokens if word not in stopwords.words("english")]

    return tokens
```

```python
text = tokenize(combined_text_high)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(stop_words = 'english')
words = cv.fit_transform(text)

sum_words = words.sum(axis=0)

words_freq = [(word, sum_words[0, i]) for word, i in cv.vocabulary_.items()]
words_freq = sorted(words_freq, key = lambda x: x[1], reverse = True)

frequency = pd.DataFrame(words_freq, columns=['word', 'freq'])

frequency.head(30).plot(x='word', y='freq', kind='bar', figsize=(15, 7), color = 'blue')
plt.title("Most Frequently Occuring Words in post with high engagement score - Top 30")
```

**The chart above displays frequently occurring words, such as 'Stanbicibtc,' representing the bank's name. Additionally, '909,' the bank's USSD code, is prominent. Words like 'Money,' 'Fund', 'Join', 'Winner' and 'Win' are also notable in the chart indicating offer from the bank. Lastly works like 'http', 'com', 'click' are also promimnet indicating the presence of links in the post**

Let us create a word cloud to see these words

```python
from wordcloud import WordCloud

wordcloud = WordCloud( width = 1000, height = 1000).generate_from_frequencies(dict(words_freq))

plt.figure(figsize=(10,8))
plt.imshow(wordcloud)
plt.axis('off')
plt.title("WordCloud - Vocabulary from Posts", fontsize = 22)
```

Great!!!
